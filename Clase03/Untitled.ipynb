```python
#Pandas el paquete de programacion para el analisis de datos
import pandas as pd
```

```python
#Origen de datos -Diccionario
data = {'Nombre' : ['Juan','Maria','Pedro'],
        'Edad':[25,30,22],
        'Ciudad':['Bogota','California','La Paz'] 
}
data['Nombre']
data['Edad']
data['Ciudad']

#Crear un DataFrame a partir de este diccionario de Python
df = pd.DataFrame(data)
df
```

```python
#Mostrar las primera filas del Dataframe
df.head()
```

```python
#Obtener Informacion sobre el dataframe
df.info()
```

```python
#Filtrar filas segun la condicion
df_mayor_25=df[df['Edad']>=25]
df_mayor_25
```
