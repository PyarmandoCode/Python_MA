```python
import openpyxl
def origen_datos_xlsx(ruta):
    try:
        libro_trabajo=openpyxl.load_workbook(ruta)
        hojas_disponibles=libro_trabajo.sheetnames
        nombre_hoja=hojas_disponibles[0]
        hoja_trabajo=libro_trabajo[nombre_hoja]
        for fila in hoja_trabajo.iter_rows(min_row=1,max_row=hoja_trabajo.max_row,values_only=True):
            print(fila)    
    except Exception as err:
        print(f"Ocurrio un error al aperturar el archivo {err}")
    else:
        print("El Archivo se apertura de manera correcta")
    finally:
        libro_trabajo.close()

origen_datos_xlsx('productod.xlsx')        
    
```

```python
for item in document.iterfind('book'):
    print(item.findtext('genre'))
```

```python
author=[]
title=[]
genre=[]
price=[]
description=[]
publish_date=[]
for item in document.iterfind('book'):
    author.append(item.findtext('author'))
    title.append(item.findtext('title'))
    genre.append(item.findtext('genre'))
    price.append(item.findtext('price'))
    description.append(item.findtext('description'))
    publish_date.append(item.findtext('publish_date'))
author    
```

```python
df=pd.DataFrame({
             'title':title,
             'author':author,
             'price':price,
             'genre':genre,
             'publish_date':publish_date
    
})
df
```

```python
df_author=df[df['author']=='Corets, Eva']
df_author
```

```python
df_author.to_csv('author_Corets',columns=['title','price','genre'],index=False)
```
