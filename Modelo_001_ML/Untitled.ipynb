```python
import tensorflow as tf
import numpy as np
```

```python
#Paso 1 Definir los datos de entrada y salida
x_train = np.array([[1.0,2.0],[2.0,3.0],[3.0,4.0],[4.0,5.0]])#datos_entrada
y_train=np.array([[3.0] ,[5.0],[7.0],[9.0]])#datos_de_salida
```

```python
#Paso 2 Crear la red Neuronal
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10,activation='relu',input_shape=(2,)),
    tf.keras.layers.Dense(1)
])
```

```python
#Paso 3 Compilar Modelo
model.compile(optimizer='adam',loss='mean_squared_error')
```

```python
#Paso4 Entrenar el Modelo
model.fit(x_train,y_train,epochs=1000)
```

```python
#Paso5 Hacer Predciones
x_test=np.array([[2.0,6.0]])
predictions=model.predict(x_test)
```

```python
print("datos entrada",x_test)
print("Prediccion de la suma",predictions[0][0])
```
