## Aprendizaje Supervizado
- **Clasificacion .-** Tiene como objetivo clasificar las entradas en un numero determinado de categorias o clases basandose en los datos etiquetados con los que fue entrenado.
- Algoritmos de Python (Modulos)
   - Redes Neuronales
   - Clasificadores de Naive Buyes
   - Vecinos Proximos (KNN) 



```python
#Ejemplo de Naive Buyes
#Supongamos que tenemos un conjunto de datos que representa mensaje de texto
#y queremos clasificarlos como spam y no spam utilizando Naive Bayes.
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
#Datos de ejemplo
mensajes =["Oferta exclusiva solo hoy, !Compra Ahora!",
           "Hola, Â¿como estas",
           "Gana un Iphone gratis.Responde a la encuesta",
           "Reunion a las 3pm en la sala de conferencias"
          ]
etiquetas =["spam","no_spam","spam","no_spam"]

#Convertir mensajes a vectores de caracteristicas
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(mensajes)

#Dividir los datos en conjuntos de entrenamiento y prueba
X_train,X_test,y_train,y_test=train_test_split(X,etiquetas,test_size=0.2,random_state=42)

#Crea el Modelo de Naive Bayes
modelo=MultinomialNB()
modelo.fit(X_train,y_train)

#hacer predicciones
predicciones=modelo.predict(X_test)

#Evaluar la precision del modelo
acurracy=accuracy_score(y_test,predicciones)
print(f"Precision del modelo: {acurracy:.2f}")

#Mostrar el informe de clasificacion
print("Informe de clasificacion")
print(classification_report(y_test,predicciones))
```

## Aprendizaje No Supervizado
 - **Agrupamiento (Clustering) .-** Reune un conjunto de algortimos de aprendizaje cuyo objetivo es agrupar los datos no etiqeutados con propiedades similares.Aislar patrones o familiar permitiendo su agrupamiento.
      - Agrupacion Jeraquica
      - Modelos Gaussianos Mixtos
      - Mapas Auto-organizados
      - Agrupacion Jerarquica

```python
#Ejemplo de Modelo Gaussiano
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
```

```python
#Generar datos de ejemplo en tres grupos
X,y_true=make_blobs(n_samples=300,centers=3,random_state=42)
#Crear un Modelo GMM con tres componentes
gmm=GaussianMixture(n_components=3,random_state=42)
#Ajustar el modelo  los datos
gmm.fit(X)
#Predecir las etiqeutas de las clases
y_pred=gmm.predict(X)
#Visualizar los resultados
centers=gmm.means_
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200, linewidths=3, label='Centros de los componentes')
plt.legend()
plt.show()
```

## Regresion Lineal
- Es un metodo estadistico que se utiliza para modelar la relacion lineal entre una variable dependiente (variable que estamos tratando de predecir) y una o mas variables independientes(las variables que utilizaremos para hacer las predecciones)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Datos de ejemplo
X=np.array([1,2,3,4,5]).reshape(-1,1)#Variable independiente(Caracteristica)
y=np.array([2,4,5,4,5]) #Variable dependiente(etiqueta)

#Crear el modelo de regresion lineal y lo entrenamos
modelo=LinearRegression()
modelo.fit(X,y)

#Hacer predicciones con nuevos datos
nuevos_datos=np.array([6,7,8]).reshape(-1,1)
predicciones=modelo.predict(nuevos_datos)
#Visualizar los datos
plt.scatter(X,y,color="black",label="Datos Reales")
plt.plot(nuevos_datos,predicciones,color="blue",marker='o',linestyle='dashed',linewidth=2,markersize=8,label='Regresion Lineal')
plt.title("Regresion Lineal Simple en Python")
plt.xlabel("Variable Independiente(X)")
plt.ylabel("Variable Dependiente(y)")
plt.legend()
plt.show()
```
